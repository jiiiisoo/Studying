**트리**

***계층적 관계***

트리: 데이터의 상,하 관계를 저장하는 자료 구조로 상,하 관계를 계층적 관계라고 한다. 예를 들면 회사의 임직원 관계, 상하 폴더 등이 있다.

이러한 계층적 데이터 관계를 저장하기 위해서는 트리가 필요하다. 배열, 링크드 리스트는 선형적 자료구조이고 해쉬함수도 계층적 데이터를 저장하기에 적합하지 않기 때문이다.

***트리란?***

트리도 링크드 리스트와 같이 노드가 있고 이를 트리 노드라고 한다. 각 노드에는 하위 노드, 즉 자식 노드를 가리키는 레퍼런스가 저장되고 이를 보통 화살표로 표시한다. 

부모 노드: 직속 상위 노드

자식 노드: 직속 하위 노드

root 노드: 가장 상위, 즉 트리의 시작 노드

형제 노드: 같은 부모를 갖는 노드

leaf(잎, 말단) 노드: 자식을 갖지 않는 노드

깊이: 특정 노드가 root 노드로부터 떨어져 있는 거리 (root 노드의 자식 노드 깊이=1)

레벨: 깊이+1, 특정 깊이를 가진 노드들을 묶어서 부를 때 사용

높이: 트리에서 가장 깊이 있는 노드의 깊이

부분 트리(sub-tree): 현재 트리의 일부분을 이루는 더 작은 트리 / 하나의 트리에 여러 부분 트리 존재하기도 함

***트리의 활용***

계층적 관계 있는 데이터를 컴퓨터에서 사용 가능

컴퓨터 과학의 다양한 문제 기발하게 해결 가능

흔히 사용하는 여러가지 추상 자료형 구현 가능: 딕셔너리, 우선순위 큐, 세트 등



***이진 트리***

각 노드가 최대 2개의 노드를 갖는 트리이다. 노드가 2개밖에 없어 왼쪽 자식, 오른쪽 자식으로 노드를 구분한다.

```python
##이진 트리 구현
class Node:
  def __init__(self,data):
    self.data=data
    self.left_child=None
    self.right_child=None

#이진 트리 노드 저장
root_node=Node(2)
node_B=Node(3)
node_C=Node(5)
node_D=Node(7)
node_E=Node(11)
root_node.left_child=node_B
root_node.right_child=node_C
node_B.left_child=node_D
node_B.right_child=node_E
```

***이진 트리 종류***

정 이진 트리 (Full Binary Tree): 모든 노드가 0 또는 2 개의 자식 노드를 갖는 트리

완전 이진 트리 (Complete Binary Tree): 마지막 노드의 직전 레벨의 노드까지 모든 노드들이 자식 노드 2개를 갖고, 마지막 레벨에서는 모든 노드들이 다 채워질 필요는 없어도 왼쪽부터 오른쪽 방향으로 모든 노드가 채워진 트리

image.png

완전 이진 트리의 노드는 트리 안에 저장된 노드가 n개일 때, log_2(n)에 비례한다.

image.png

따라서, 이진 트리의 높이를 h, 저장된 노드의 개수를 n이라고 하면, 2^h<=n<2^(h+1) 이 성립하므로, log_2를 각 항에 씌워주면 h<=log_2(n)<h+1 이므로, 노드의 높이는 log_2(n)을 넘지 않는 최대의 정수임을 알 수 있다. 그러므로 노드의 높이는 log_2(n)에 비례한다. 또, 높이가 1만큼 증가하려면 최소한 현재 노드보다 2배 이상이 되어야 한다는 것도 알 수 있다.

포화 이진 트리 (Perfect Binary Tree): 모든 레벨이 빠짐없이 다 노드로 채워진 이진 트리

따라서, n=2^(h+1)-1 으로 노드에 따른 높이가 고정된다.

***완전 이진 트리 구현***

완전 이진 트리는 리스트로 구현할 수 있다.

image.png

자식 노드를 찾는 방법: 왼쪽 자식 노드의 경우 (부모노드인덱스 ⋅ 2)를 하고, 오른쪽 자식 노드의 경우 (부모노드인덱스 ⋅ 2)+1을 한다.

부모 노드를 찾는 방법: 왼쪽 자식 노드에서 찾을 경우 (자식노드/2)를 하고, 오른쪽 자식 노드에서 찾을 경우 (자식노드/2)를 한 뒤 나오는 정수값만 취하면 된다.

```python
complete_binary_tree=[None,1,5,12,11,9,10,14,2,10] #None 넣는 이유: 파이썬 리스트는 0부터 시작하므로

def get_parent_index(complete_binary_tree, index):
    if index//2>len(complete_binary_tree)-1 or index//2==0:
        return None
    else:
        return index//2

def get_left_child_index(complete_binary_tree, index):
    if index*2>len(complete_binary_tree)-1:
        return None
    else:
        return index*2

def get_right_child_index(complete_binary_tree, index):
    if index*2+1>len(complete_binary_tree)-1:
        return None
    else:
        return index*2+1
```

***트리 순회***

순회: 자료 구조에 저장된 모든 데이터를 출력하는 것

트리에는 선형적 관계가 없으므로 어떤 순서로 출력할 지가 중요함

이 때, 트리 출력에는 반복문이 아닌 재귀함수가 사용되고 트리에서는 거의 재귀함수를 사용한다.

순회 기본 동작: 재귀적으로 왼쪽 부분 트리 순회 / 재귀적으로 오른쪽 부분 트리 순회 / 현재 노드 데이터 출력

pre-order 순회: 현재 노드 출력->재귀적으로 왼쪽 부분 트리 순회->재귀적으로 오른쪽 부분 트리 순회

post-order 순회: 재귀적으로 왼쪽 부분 트리 순회-> 재귀적으로 오른쪽 부분 트리 순회-> 현재 노드 출력

image.png

in-order 순회: 재귀적으로 왼쪽 부분 트리 순회->현재 노드 출력->재귀적으로 오른쪽 부분 트리 순회

image.png

이렇듯, 순회를 사용하면 계층적 관계의 데이터를 선형적 관계로 나타낼 수 있다.



```python
#in-order 순회
def traverse_inorder(node):
  if node!=None:
    traverse_inorder(node.left_child)
    print(node.data)
    traverse_inorder(node.right_child)

# 여러 노드 인스턴스 생성
node_A = Node("A")
node_B = Node("B")
node_C = Node("C")
node_D = Node("D")
node_E = Node("E")
node_F = Node("F")
node_G = Node("G")
node_H = Node("H")
node_I = Node("I")

# 생성한 노드 인스턴스들 연결
node_F.left_child = node_B
node_F.right_child = node_G

node_B.left_child = node_A
node_B.right_child = node_D

node_D.left_child = node_C
node_D.right_child = node_E

node_G.right_child = node_I

node_I.left_child = node_H

# 노드 F를 root 노드로 만든다
root_node = node_F

# 만들어 놓은 트리를 in-order로 순회한다
traverse_inorder(root_node)
```
