***기능 vs. 구현***

기능: 연산이 무엇을 하는 지

구현: 연산의 기능을 어떻게 하는 지

***추상화***

추상화는 기능만 알고 구현은 몰라도 되는 것이다. 예를 들어, 어떤 한 함수가 정의되어 있다면 그 함수의 기능만 알면 구현은 몰라도 쓸 수 있다. 따라서 개발자들이 서로 함수들을 공유하고 이용하면 훨씬 편리하게 일을 해결할 수 있다.

***추상 자료형***

특징

데이터 사이 순서 관계를 유지할 수 있다

접근 연산: 특정 위치에 있는 데이터를 가지고 오거나 수정할 수 있다
탐색 연산: 특정 조건을 만족하는 데이터를 찾을 수 있다
삽입 연산: 특정 위치에 새로운 데이터를 저장해줄 수 있다
삭제 연산: 특정 위치에 데이터를 지울 수 있다

이런 특징, 연산 가지는 존재를 리스트라고 한다.

따라서 구현은 모르고 기능만 알면 될 때는 추상 자료형을, 최적화 시킬 때 (속도 더 빠르게 등)을 할 때는 자료 구주형에 더 집중하면 된다.

***리스트 개념***

데이터 간 순서 관계를 유지할 수 있다. 탐색 연산, 삭제 연산, 접근 연산, 삽입 연산이 가능하다.

***리스트 구현***

리스트에는 크게 동적 배열, 링크드 리스트가 있다. 둘 중 무엇을 쓸 지는 써야 하는 기능의 시간 복잡도를 비교하면 된다.

파이썬의 리스트는 동적배열이므로 시간복잡도를 알고 싶으면 동적배열의 시간복잡도를 보면 된다.

큐Queue
데이터를 삭제 시 앞에서만. 뒤에서만 삽입
FIFO First In First out
맨 뒤 데이터 추가, 앞 데이ㅓ 삭제, 앞 데이터 접근
데이터 간 순서 
deque doubly-ended queue 
맨 앞과 뒤에 데이터 삽입,삭제하 ㄹ 수 있게 해줌 

삭제하고 삭제 한 데이터 return 

```python
from collections import deque
queue=deque()
queue.qppend()
queue[0]
queue.popleft()
```
